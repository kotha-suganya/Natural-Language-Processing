**Natural Language Processing**

LEMMATIZATION

```python
# Lemmatization
import nltk
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
text="The cats were running faster than the dogs."
tokens=word_tokenize(text)
lemmatized_tokens=[lemmatizer.lemmatize(token) for token in tokens]
print(f"Original Text: {text}")
print(f"Lemmatized words: {lemmatized_tokens}")
```

STOPWORDS

```python
# stopwords
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
text = "Hello students! This is a simple NLP mini project. Let us now start to learn how tokenization works."
text=text.lower()
text=text.translate(str.maketrans('','',string.punctuation))
tokens = word_tokenize(text)
filtered=[word for word in tokens if word not in stopwords.words('english')]
print(f"tokens: {tokens}")
print(f"filtered words: {filtered}")
```

WORD COUNT

```python
# word count
from collections import Counter
from nltk.tokenize import word_tokenize
text = "NLP is fun. NLP is powerful.Learning NLP is useful."
tokens = word_tokenize(text)
word_counts = Counter(tokens)
print(f"word counts: {word_counts}")
print("Most common:", word_counts.most_common(2))
```

Named Entity Recognition

```python
import spacy
nlp=spacy.load('en_core_web_sm')
text="Elon Musk founded SpaceX in the United States."
doc=nlp(text)
for ent in doc.ents:
  print(ent.text,"-",ent.label_)
```

SPAM EMAIL CLASSIFIER

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
texts=["Free money now","Hi how are you","win cash price","Let's meet tomorrow"]
lables=[1,0,1,0]    # 1 for spam and 0 for not spam
vectorizer=CountVectorizer()
x=vectorizer.fit_transform(texts)
model=MultinomialNB()
model.fit(x,lables)
text=["You got 50 likes"]
test_vector=vectorizer.transform(text)
prediction=model.predict(test_vector)
print(f"email classified as: {prediction}")
```

TEXT SIMILARITY CHECKER

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
texts=["I love machine learning","I like AI and machine learning"]
vectorizer=TfidfVectorizer()
tfidf=vectorizer.fit_transform(texts)
similarity=cosine_similarity(tfidf[0:1],tfidf[1:2])
print("Similarity Score:",similarity[0][0])
```

MINIMUM EDIT DISTANCE

```python
import nltk
from nltk.metrics import edit_distance
word1="kitten"
word2="sitting"
distance=edit_distance(word1,word2)
print("Minimum Edit Distance:",distance)
```

N-GRAM ALGORITHM

```python
from nltk.util import ngrams
sentence="I love NLP"
tokens=sentence.split()
bigrams=list(ngrams(tokens,2))
trigrams=list(ngrams(tokens,3))
print("Bigrams:",bigrams)
print("Trigrams:",trigrams)
```

STEMMING ALGORITHM


```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
words=["playing","played","plays","pneumonoultramicroscopicsilicovolcanoconiosis"]
for w in words:
  print(stemmer.stem(w))
```

MAXIMUM ENTROPY MODEL ALGORITHM

```python
import nltk
from nltk.classify import MaxentClassifier
train_data=[({"word": "good"},"Positive"),
 ({"word": "excellent"},"Positive"),
            ({"word": "bad"}, "Negative"),
            ({"word": "poor"}, "Negative")]
classifier = MaxentClassifier.train(train_data, max_iter = 10)
print(classifier.classify({"word": "good"}))
print(classifier.classify({"word": "bad"}))
```

HIDDEN MARKOV MODEL

```python
import nltk
from nltk.tag import hmm
train_data = [
    [('I', 'PRON'), ('eat', 'VERB'), ('pizza', 'NOUN')],
    [('He', 'PRON'), ('like', 'VERB'), ('pizza', 'NOUN')]
]
tagger = hmm.HiddenMarkovModelTrainer()
model = tagger.train(train_data)
test_sentence = [('I'), ('like'), ('pizza')]
print(model.tag(test_sentence))
```
